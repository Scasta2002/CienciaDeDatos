**P5.** Filtra el DataFrame para quedarte solo con las flores de la especie `virginica`.

```python
# Tu respuesta aquí
df_virginica = df[df['species'] == 'virginica'][['sepal_length', 'species']]
print("Del tipo de flores virginica hay", len(df_virginica), "filas")
print(df_virginica)

```

**P6.** Filtra las filas donde `petal_length` sea mayor a 5.0. Imprime cuántas filas cumplen esa condición.

```python
# Tu respuesta aquí
df_virginica_mayores05 = df[
    (df['species'] == 'virginica') & 
    (df['petal_length'] > 5.0)
][['sepal_length', 'species']]

print("Del tipo de flores virginica y con longitud mayor a 5.0 hay", len(df_virginica_mayores05), "filas")
print(df_virginica_mayores05)
```
